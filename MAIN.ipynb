```python
if not os.path.exists("dan"):
    print("[ERROR] there is no directory 'dan'")
    sys.exit()
known_encodings = []
images = os.listdir("dan")
# print(images)
face_enc_dan = []
for(i, image) in enumerate(images):
    print(f"[+] processing img {i + 1}/{len(images)}")
    print(image)
    face_img = face_recognition.load_image_file(f"dan/{image}")
    face_enc_dan = face_recognition.face_encodings(face_img)[0]
    if len(known_encodings) == 0:
        known_encodings.append(face_enc_dan)
    else:
        for item in range(0, len(known_encodings)):
            result = face_recognition.compare_faces([face_enc_dan], known_encodings[item], tolerance=0.7)
            if result[0]:
                known_encodings.append(face_enc_dan)
                print("Same person!")
                break
            else:
                print("Another person!")
                break


```
